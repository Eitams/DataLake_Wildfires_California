### 1. Wildfire database 

```python
try: 
    print("dbname={}".format(DB_NAME_FETCH))
    conn = psycopg2.connect("host={} dbname={} user={} password={}".format(
            ENDPOINT_FETCH, DB_NAME_FETCH, USERNAME_FETCH, PASSWORD_FETCH))    
except psycopg2.Error as e: 
    print("Error: Could not make connection to the Postgres database")
    print(e)
```

```python
try: 
    cur = conn.cursor()
except psycopg2.Error as e: 
    print("Error: Could not get curser to the Database")
    print(e)
    
# Auto commit is very important
conn.set_session(autocommit=True)
```

```python
try: 
    cur.execute("SELECT FireDiscoveryDateTime, InitialLatitude, InitialLongitude FROM Firedata_Total;")
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
data=[]
while row:
    data.append(list(row))
    row = cur.fetchone()

header=['date','lat','long']
df=pd.DataFrame(data, columns=header)
df
```

```python
cur.close()
conn.close()
```

### 2. Weather database

```python
try: 
    print("dbname={}".format(DB_NAME))
    conn = psycopg2.connect("host={} dbname={} user={} password={}".format(ENDPOINT, DB_NAME, USER, PASSWORD))
except psycopg2.Error as e: 
    print("Error: Could not make connection to the Postgres database")
    print(e)
```

```python
try:
    cur = conn.cursor()
except psycopg2.Error as e: 
    print("Error: Could not get curser to the Database")
    print(e)
    
conn.set_session(autocommit=True)
```

#### a. Historical Weather data

```python
try: 
    cur.execute("SELECT * FROM HISTORICAL_WEATHER_DATA;")
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)
row = cur.fetchone()
data=[]
while row:
    data.append(list(row))
    row = cur.fetchone()

header=['id' , 'address' , 'name' , 'latitude' , 'longitude' , 'tz' , 'wdir' , 'temp' , 'maxt' , 'visibility' ,'spd' , 'datetimeStr' , 'solarenergy' , 'heatindex' , 'cloudcover' , 'mint' , 'precip' , 'solarradiation' , 'weathertype' ,	'snowdepth' , 'sealevelpressure' , 'snow' , 'dew' , 'humidity' , 'precipcover' , 'wgust' , 'conditions' , 'windchill']
df=pd.DataFrame(data, columns=header)
new_df=df
new_df.to_csv('../data/HISTORICAL_WEATHER_DATA.csv', index=False)
df.shape
```

```python
#If you would like to delete the data in the table
cur.execute('TRUNCATE TABLE HISTORICAL_WEATHER_DATA')
```
